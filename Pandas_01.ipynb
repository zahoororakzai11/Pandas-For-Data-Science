## Pandas Series


```python
import pandas as pd
```

#### Creating a Series

```python
# To Creat a series
numbers = range(1,100,5)
pd.Series(numbers)
```

```python
# Creat a object series
string = "Hi","How","are ","you","?"
pd.Series(string)
```

```python
# Create a series by having numeric and string values
a = pd.Series([12,'Peshawar',122.3,"Happy Birth day"])
a  # Numberic values are treated as object
```

```python
# To Set index value for Series
marks = [10,20,30,40,50]
subject = ['Maths','Urdu','English','Arabic','Chemistry']
pd.Series(subject,index = marks)
```

```python
# Creat a Series from Dictionary

data =  {'Maths': 60, 'Science': 89, 'English': 76, 'Social Science': 86}
pd.Series(data)
```

```python
# A Series with missing values
subjects = ["Maths", "Science", "Art and Craft" , "Social Science"]
marks_series = pd.Series(data,index = subjects)
marks_series
```

#### 3.2 Manipulating Series

#### 6. Sorting a numeric series

```python
# create a pandas series
import pandas as pd
import numpy as np
values = pd.Series([23,45,np.nan,41,23,34,55,np.nan,34,20])
values.sort_values(ascending = 1)
```

```python
# Sorting a Categorial Series
string_values = pd.Series(['a','b','f','h','z','g','m','d'])
string_values.sort_values(ascending=True)
```

```python
# Rank a Series
marks_series.rank(ascending=True,pct= False)
```

### Pandas DataFrame

```python
# Create A DataFrame
data = {
    'Subject':["English","History","Science","English","Georaphy","Art"],
    'Marks':(45,50,60,56,52,80),
    'CGPA':[2.5,3.0,3.5,2.0,4.0,4.0]}
df = pd.DataFrame(data)
df
```

```python
# To Create a DataFrame from Series
Subject = pd.Series(["Maths","History","Science","English","Georaphy"])
Marks = pd.Series([45,65,78,80,65,78])
CGPA = pd.Series([2.5,3.0,3.5,2.0,4.0,4.0])
df = pd.DataFrame([Subject,Marks,CGPA],index = ['Subject','Marks','CGPA']).T # here use T for Vertical transpose
df
```

```python
# Create DataFrame from List
Subject = ['Maths', 'History', 'Science', 'English', 'Georaphy', 'Art']
Marks = [45, 65, 78, 65, 80, 78]
CGPA = [2.5, 3.0, 3.5, 2.0, 4.0, 4.0]
pd.DataFrame([Subject,Marks,CGPA],index = ['Subject','Marks','CGPA'])
```
